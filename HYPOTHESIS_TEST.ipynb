```python
import pandas as pd
import scipy.stats as stats
import numpy as np
```

```python
df = pd.read_csv('/content/Hypothesis_testing dataset.csv')
```

```python
df.head()
```

Overall Insight:
The hypothesis tests indicate that:

*Students' sugary drink consumption is not significantly greater than 4 drinks per week.

*The proportion of students frequently eating fast food is not significantly higher than 60%.

*Thus, based on the sample data and α = 0.05, the study does not support the claims that students consume excessive sugary drinks or that a large majority frequently eat fast food.

 1. Test for Sugary Drinks Mean

 H0: μ = 4

 H1: μ > 4

```python
sugary_drinks = df["Sugary_Drinks_Per_Week"]
mu0 = 4
```

One sample t-test

```python
t_stat, p_value = stats.ttest_1samp(sugary_drinks, mu0)
```

Convert to one-tailed

```python
p_value_one_tail = p_value / 2

print("Sugary Drinks Hypothesis Test")
print("t-statistic:", t_stat)
print("p-value (one-tailed):", p_value_one_tail)

if p_value_one_tail < 0.05 and t_stat > 0:
    print("Reject H0: Students drink more than 4 sugary drinks per week")
else:
    print("Fail to Reject H0")
```

2. Test for Fast Food Proportion

 H0: p = 0.60

 H1: p > 0.60

```python
fast_food = df["FastFood_Per_Week"]
```

 Students eating fast food >=3 times/week

```python
count = np.sum(fast_food >= 3)
n = len(fast_food)

p_hat = count / n
p0 = 0.60
```

z-test for proportion

```python
z = (p_hat - p0) / np.sqrt((p0 * (1 - p0)) / n)
```

One-tailed p-value

```python
p_value_prop = 1 - stats.norm.cdf(z)
print("\nFast Food Hypothesis Test")
print("Sample proportion:", p_hat)
print("z-statistic:", z)
print("p-value:", p_value_prop)

if p_value_prop < 0.05:
    print("Reject H0: More than 60% eat fast food >=3 times/week")
else:
    print("Fail to Reject H0")
```
